```python
import numpy as np
import matplotlib.pyplot as plt
import cmath as cm
```

```python
from matplotlib.pyplot import figure

figure(figsize=(15, 15))
```

```python
def f(z):
    return np.exp(1/z)
```

```python
x = np.array([complex(m,t) for t in range(-1000,1000) for m in range(-1000,1000)])
y = np.array([complex(t,m) for t in range(-1000,1000) for m in range (-1000,1000)])
z1 = f(x)
z2 = f(y)
```

```python
# plt.figure(figsize=(15,15))
# plt.plot(x.real,x.imag )
# plt.plot(y.real,y.imag)
# plt.show()
plt.figure(figsize=(15,15))
plt.scatter(z1.real,z1.imag,s=0.5)
plt.scatter(z2.real,z2.imag,s=0.3)
```
